```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
import numpy as np
```

```python
df = pd.read_csv("../datavis/final_table.csv")
```

```python
all_data = df[['MCC_VISA_report_name', 'anon_Kwota', 'LocCity']]
one_hot_cat = pd.get_dummies(all_data['MCC_VISA_report_name'])
one_hot_city = pd.get_dummies(all_data['LocCity'])
all_data = all_data.drop('MCC_VISA_report_name',axis = 1)
all_data = all_data.drop('LocCity',axis = 1)
# Join the encoded df
all_data = all_data.join(one_hot_cat)
all_data = all_data.join(one_hot_city)
```

```python
df[['MCC_VISA_report_name', 'anon_Kwota', 'LocCity']].groupby(['MCC_VISA_report_name', 'LocCity']).mean()
```

```python
X_train = all_data.drop('anon_Kwota', axis = 1)
y_train = all_data['anon_Kwota']
```

```python
regr = svm.SVR()
regr.fit(X_train, y_train)
```

```python
def transform_input(category, area):
    model_input = {
        'Building Materials & Services':0,
        'Business & Personal Services':0,
        'Clothing':0,
        'Entertainment':0,
        'Financial Services':0,
        'Food & Drink':0,
        'General Retail & High Street':0,
        'Health':0,
        'Motoring':0,
        'Petrol':0,
        'Restaurants':0,
        'Services':0,
        'Supermarket':0,
        'Training & Education':0,
        'Travel':0,
        'Wholesale':0,
        'WARSZAWA':0,
        'Warszawa (Bemowo)':0,
        'Warszawa (Białołęka)':0,
        'Warszawa (Bielany)':0,
        'Warszawa (Mokotów)':0,
        'Warszawa (Ochota)':0,
        'Warszawa (Praga-Południe)':0,
        'Warszawa (Praga-Północ)':0,
        'Warszawa (Rembertów)':0,
        'Warszawa (Targówek)':0,
        'Warszawa (Ursus)':0,
        'Warszawa (Ursynów)':0,
        'Warszawa (Wawer)':0,
        'Warszawa (Wesoła)':0,
        'Warszawa (Wilanów)':0,
        'Warszawa (Wola)':0,
        'Warszawa (Włochy)':0,
        'Warszawa (Śródmieście)':0,
        'Warszawa (Żoliborz)':0,
    }
    model_input[category] = 1
    model_input[area] = 1
    
    return pd.DataFrame(pd.Series(model_input)).transpose()
```

```python
regr.predict(transform_input('Financial Services', 'Warszawa (Ursynów)'))
```
